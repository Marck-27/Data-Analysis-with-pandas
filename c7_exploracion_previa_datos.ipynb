### Algunas funciones usadas en este script:
`.head(N)` <- Muestra los primeros N registros de un data frame 

`.tail(N)` <- Muestra los últimos N registros de un data frame 

`.sample(N)` <- Muestra N registros aleatorios de un data frame 

`.shape` <- Extrae numero de renglones y columnas de un data frame 

`.info()` <- Muestra la información del data frame (por ejemplo número de campos no nulos)

`.describe()` <- Muestra información estadistica del data frame

`.describe(include='all')` <- Muestra más información estadistica del data frame (incluye frecuencas y clases de datos)

`.dtypes` <- Muestra el tipo de dato de cada variable

`.convert_dtypes().dtypes` <- Convierte los tipos de datos de un dataframe a tipos de datos optimizados por pandas

```python
import pandas as pd;
```

```python
df_meteorites = pd.read_csv('./db/NASA/Meteorite_Landings.csv')
```

```python
# Visualizamos los primeros N registros:
df_meteorites.head(10)
```

```python
# Visualizamos los últimos N registros:
df_meteorites.tail(10)
```

```python
# Visualizamos N registros aleatorios:
df_meteorites.sample(10)
```

```python
# Numero de renglones y columnas del dataframe
df_meteorites.shape
```

```python
# Extraemos información del dataframe
df_meteorites.info()
```

```python
# Extraemos información estadistica del dataframe
df_meteorites.describe()
```

```python
# Extraemos MÁS información estadistica del dataframe
# Incluyendo las clases de datos ('unique') y frecuencias
df_meteorites.describe(include='all')
```

```python
# Muestra el tipo de datos en cada variable (columna)
df_meteorites.dtypes
```

```python
# Convierte los tipos de datos a tipos más optimizados para pandas
df_meteorites.convert_dtypes().dtypes
```
